This is a code to build a email classifier which learns from emails distinct pattern and classify it either postive or negative.
We build multiple  models for it and choose the best one

```python
# Natural Language Processing

# Importing the libraries
import pandas as pd
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.metrics import roc_curve, auc

# Importing the dataset
dataset = pd.read_csv('Opportunities - Sheet1.tsv', delimiter = '\t')
```

```python
dataset['Replies'].tail()
```

```python
# Cleaning the texts
import re
from nltk.stem.snowball import SnowballStemmer

```

```python

#Unpickle stopwords
with open("Stopwords.pickle","rb") as stop_file:
    stop = pkl.load(stop_file)
    
```

```python

stop
```

```python

stop = stop - set(('yes','no','not','don\'t','aren\'t','isn\'t','doesn\'t','isn','aren','don','doesn'))
```

```python

ss = SnowballStemmer("english")
```

```python
corpus = []
for i in range(0,len(dataset)):
    review = re.sub('[^a-zA-Z]', ' ', dataset['Replies'][i])
    review = review.lower()
    review = review.split()
    review = [ss.stem(word) for word in review if not word in set(stop)]
    review = ' '.join(review)
    corpus.append(review)
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, 1].values
feature_names = cv.get_feature_names()
```

```python
feature_names[300:400] # preview some features
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection  import train_test_split
from sklearn.metrics import confusion_matrix,roc_curve, auc
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)

from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, VotingClassifier

######################Model Building######################
# Fitting Naive Bayes 
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(X_train, y_train)
y_pred_log2 = classifier_nb.predict(X_test)
cm_log2 = confusion_matrix(y_test, y_pred_log2)
#a1= classifier_nb.score(X_train,y_train)
#b1=classifier_nb.score(X_test,y_test)
print(cm_log2)

#ROC Curve
y_prob_nb = classifier_nb.predict_proba(X_test)[:,1] # This will give you positive class prediction probabilities  
y_pred_nb = np.where(y_prob_nb > 0.5, 1, 0) # This will threshold the probabilities to give class predictions.

X_test.shape[0]
(y_test!= y_pred_nb).sum()

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_prob_nb)
roc_auc = auc(false_positive_rate, true_positive_rate)

plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='blue',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
```

```python
#Fitting SVM
from sklearn.svm import SVC
classifier_svm = SVC()
classifier_svm.fit(X_train, y_train)
y_pred_log3 = classifier_svm.predict(X_test)
cm_log3 = confusion_matrix(y_test, y_pred_log3)
print (cm_log3)
#a2=classifier_ksvm.score(X_train,y_train)


print (classifier_svm.score(X_test,y_test))

#ROC Curve
y_prob_svm = classifier_svm.predict(X_test) # This will give you positive class prediction probabilities  
y_pred_svm = np.where(y_prob_svm > 0.5, 1, 0) # This will threshold the probabilities to give class predictions.

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_prob_svm)
roc_auc = auc(false_positive_rate, true_positive_rate)

plt.figure(figsize=(10,10))
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='blue',label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')

```
